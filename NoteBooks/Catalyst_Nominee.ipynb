```python
# library import
import sys
sys.path.append('./HR_Library/') 
import datetime as dt
import HR_Analytics as hr
import pandas as pd
```

```python
# init db engine
engine = hr.set_engine('credentials','HRODS')
```

```python
# get flat data
catalyst_data = hr.file_query("Nominee List.xlsx", "PD")
catalyst_data
```

```python
# query
query = """
SELECT 
        "Ethnicity",
        Count(*) AS "HC"
    FROM 
        "HPW_DATA"
    WHERE 
        "Report Date" = '{0}'
    AND 
        "Worker Reg / Temp Code" = 'R'
    AND 
        "Worker Status Category Code" = 'A'
    AND
        "Pay Group Country Code" = 'USA'
    AND 
        "Management Level Category" IN ('EXEC')
    GROUP BY 
        "Ethnicity"
"""
```

```python
# report date
data_date = '2019-10-31'
# run query
data = hr.internal_query(query.format(data_date), engine)

# process data
data['%']  = round((data['HC'] / data['HC'].sum()) * 100,1).astype(str) + '%'
data.sort_values(by='Ethnicity',ascending=False)
```
